```python
#Remove Emojis

def emoji(msg): 
    
    msg_emoj=msg.str.replace(r'[(\U0001F600-\U0001F92F|\U0001F300-\U0001F5FF|\U0001F680-\U0001F6FF|\U0001F190-\U0001F1FF|\U00002702-\U000027B0|\U0001F926-\U0001FA9F|\u200d|\u2640-\u2642|\u2600-\u2B55|\u23cf|\u23e9|\u231a|\ufe0f)]+','')
    
    msg_digit=msg_emoj.str.replace(r'[0-9]', ' ')

    msg_Spac = msg_digit.str.replace(r'_',' ')

    msg_Spac = msg_Spac.str.replace(r'.','')

    msg_Spac = msg_Spac.str.replace(r'!','')

    msg_Spac = msg_Spac.str.replace(r'#','')

    msg_Spac = msg_Spac.str.replace(r'%','')

    msg_Spac = msg_Spac.str.replace(r'&','')

    msg_Spac = msg_Spac.str.replace(r'’','')

    msg_Spac = msg_Spac.str.replace(r'(','')

    msg_Spac = msg_Spac.str.replace(r')','')

    msg_Spac = msg_Spac.str.replace(r'-','')

    msg_Spac = msg_Spac.str.replace(r'/','')

    msg_Spac = msg_Spac.str.replace(r':','')

    msg_Spac = msg_Spac.str.replace(r';','')

    msg_Spac = msg_Spac.str.replace(r'<','')

    msg_Spac = msg_Spac.str.replace(r'=','')

    msg_Spac = msg_Spac.str.replace(r'>','')

    msg_Spac = msg_Spac.str.replace(r'?','')

    msg_Spac = msg_Spac.str.replace(r'@','')

    msg_Spac = msg_Spac.str.replace(r'[','')

    msg_Spac = msg_Spac.str.replace(r']','')

    msg_Spac = msg_Spac.str.replace(r'^','')

    msg_Spac = msg_Spac.str.replace(r'{','')

    msg_Spac = msg_Spac.str.replace(r'}','')

    msg_Spac = msg_Spac.str.replace(r'|','')

    msg_Spac = msg_Spac.str.replace(r'~','')

    msg_Spac = msg_Spac.str.replace(r'+','')

    msg_Spac = msg_Spac.str.replace(r'\s+', ' ')


    return msg_Spac
    
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
def trainer(model_obj, param, X, X_char, train, Xt, Xt_char, Dev, Y, y_Dev):
   
    model_obj.fit(X,Y)
#Training Classifier with TF-IDF Features Word n-gram
    print("Accuracy of Word n-gram(1,2) :", (model_obj.score(Xt, y_Dev)))
#Printing Classification Report
    predict=model_obj.predict(Xt)
    print(classification_report(y_Dev, predict))
#Printing Confusion Matrix
    conf_mat=confusion_matrix(y_Dev, predict)
    print("Confusion Matrix")
    print(conf_mat)
    print(" ")
    word=model_obj.predict(test_Mal_Hasoc_word)
    
    pre=precision_score(y_Dev, predict, average='micro')
    rec=recall_score(y_Dev, predict, average='micro')
    f1s=f1_score(y_Dev, predict, average='micro')
    
    print("Precision : ", pre)
    print("Recall : ", rec)
    print("F1 Score : ", f1s)

#Training Classifier with TF-IDF Features Char n-gram
    model_obj.fit(X_char, Y)
    print("Accuracy of Char n-gram(1,5) :", (model_obj.score(Xt_char, y_Dev)))
#Printing Classification Report
    predict=model_obj.predict(Xt_char)
    print(classification_report(y_Dev, predict))
#Printing Confusion Matrix
    conf_mat=confusion_matrix(y_Dev, predict)
    print("Confusion Matrix")
    print(conf_mat)
    print(" ")
    char=model_obj.predict(test_Mal_Hasoc_char)
    
    pre=precision_score(y_Dev, predict, average='micro')
    rec=recall_score(y_Dev, predict, average='micro')
    f1s=f1_score(y_Dev, predict, average='micro')
    
    print("Precision : ", pre)
    print("Recall : ", rec)
    print("F1 Score : ", f1s)
        
              
#Training Classifier with TF-IDF Features Combine Word n-gram & Char n-gram
    model_obj.fit(train, Y)
    print("Accuracy of Combinining word & Char n-gram", (model_obj.score(Dev, y_Dev)))
#Printing Classification Report
    predict=model_obj.predict(Dev)
    print(classification_report(y_Dev, predict))
#Printing Confusion Matrix
    conf_mat=confusion_matrix(y_Dev, predict)
    print("Confusion Matrix")
    print(conf_mat)
    print(" ")
    word_char=model_obj.predict(test_Mal_Hasoc_word_char)
    
    pre=precision_score(y_Dev, predict,  average='micro')
    rec=recall_score(y_Dev, predict, average='micro')
    f1s=f1_score(y_Dev, predict, average='micro')
    
    print("Precision : ", pre,)
    print("Recall : ", rec)
    print("F1 Score : ", f1s)

    
    return word, char, word_char
```

```python
def trainer1111(model_obj, param, x_smote_Word,y_smote_Word, x_smote_Char,y_smote_Char, x_smote_Comb ,y_smote_Comb, x_smote_Word_Dev, y_smote_Word_Dev, x_smote_Char_Dev,y_smote_Char_Dev, x_smote_Comb_Dev, y_smote_Comb_Dev):
    # Training Classifier  with TF-IDF Features Word n-gram(1,2)
    GSCV=GridSearchCV(model_obj, param , cv=10,return_train_score=False)
    GSCV.fit(x_smote_Word, y_smote_Word)
    print("Accuracy of Word n-gram(1,1) :", (GSCV.score(x_smote_Word_Dev, y_smote_Word_Dev)))
    #Printing Classification Report
    predict=GSCV.predict(x_smote_Word_Dev)
    print(classification_report(y_smote_Word_Dev, predict))
    #Printing Confusion Matrix
    conf_mat=confusion_matrix(y_smote_Word_Dev, predict)
    print("Best Parameters ", GSCV.best_params_)
    print("Best Score ", GSCV.best_score_)
    print(" ")
    print("Confusion Matrix")
    print(conf_mat)
    print(" ")
    word=GSCV.predict(test_Mal_Hasoc_word)
 
    accuracy = accuracy_score(y_smote_Word_Dev, predict)
    recall = recall_score(y_smote_Word_Dev, predict, average='micro')
    precision = precision_score(y_smote_Word_Dev, predict, average='micro')
    f1score = f1_score(y_smote_Word_Dev, predict, average='micro')
    print ('accuracy: {}, recall: {}, precision: {}, f1-score: {}'.format(accuracy, recall, precision, f1score))
    
    
    #Training Classifier with TF-IDF Features Char n-gram(1,5)
    GSCV.fit(x_smote_Char, y_smote_Char)
    print("Accuracy of Char n-gram(1,5) :", (GSCV.score(x_smote_Char_Dev, y_smote_Char_Dev)))
    predict=GSCV.predict(x_smote_Char_Dev)
    print(classification_report(y_smote_Char_Dev, predict))
   #Printing Confusion Matrix
    conf_mat=confusion_matrix(y_smote_Char_Dev, predict)
    print("Best Parameters ", GSCV.best_params_)
    print("Best Score ", GSCV.best_score_)
    print(" ")
    print("Confusion Matrix")
    print(conf_mat)
    print(" ")
    char=GSCV.predict(test_Mal_Hasoc_char)
    
    accuracy = accuracy_score(y_smote_Char_Dev, predict)
    recall = recall_score(y_smote_Char_Dev, predict, average='micro')
    precision = precision_score(y_smote_Char_Dev, predict, average='micro')
    f1score = f1_score(y_smote_Char_Dev, predict, average='micro')
    print ('accuracy: {}, recall: {}, precision: {}, f1-score: {}'.format(accuracy, recall, precision, f1score))
    
    
   
    
    
    #Training Classifier with TF-IDF Features Combine Word n-gram & Char n-gram
    GSCV.fit(x_smote_Comb, y_smote_Comb)
    print("Accuracy of Combinining word & Char n-gram", (GSCV.score(x_smote_Comb, y_smote_Comb)))
    #Printing Classification Report
    predict=GSCV.predict(x_smote_Comb_Dev)
    print(classification_report(y_smote_Comb_Dev, predict))
    #Printing Confusion Matrix
    conf_mat=confusion_matrix(y_smote_Comb_Dev, predict)
    print("Best Parameters ", GSCV.best_params_)
    print("Best Score ", GSCV.best_score_)
    print(" ")
    print("Confusion Matrix")
    print(conf_mat)
    print(" ")
    word_char=GSCV.predict(test_Mal_Hasoc_word_char)
    
    accuracy = accuracy_score(y_smote_Comb_Dev, predict)
    recall = recall_score(y_smote_Comb_Dev, predict, average='micro')
    precision = precision_score(y_smote_Comb_Dev, predict, average='micro')
    f1score = f1_score(y_smote_Comb_Dev, predict, average='micro')
    print ('accuracy: {}, recall: {}, precision: {}, f1-score: {}'.format(accuracy, recall, precision, f1score))
    
    return word, char, word_char
```
